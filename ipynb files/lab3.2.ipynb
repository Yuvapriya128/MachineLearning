```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df=pd.read_csv('Salary_dataset.csv')
x=df['YearsExperience'].values.T
y=df['Salary'].values.T

n=np.size(x)
meanx=np.mean(x)
meany=np.mean(y)
Sxy=np.sum(y*x)-n*meanx*meany
Sxx=np.sum(x*x)-n*meanx*meanx
a=Sxy/Sxx#slope=ax+b
b=meany-meanx*a#intercept

print(f"Co-efficient is {a,b}")    
plt.scatter(x,y,color="m",marker="o",s=30)
ypred=(a*x)+b
plt.plot(x,ypred,color="g")
plt.xlabel('x')
plt.ylabel('y')

sse = np.sum((y - ypred) ** 2)  # Sum of Squared Errors
mse = sse / len(y)  # Mean Squared Error
rmse = np.sqrt(mse)  # Root Mean Squared Error
mae= np.mean(np.abs(y - ypred))  # Mean Absolute Error
sst = np.sum((y - meany) ** 2)  # Total Sum of Squares
ssr = sst - sse  # Sum of Squares for Regression

print(f"sse {sse},mse {mse},rmse {rmse}\
mae {mae} , sse {sst}, ssr {ssr}");


  
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hardcoded dataset
X = np.array([1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.7, 3.9, 4.0, 4.5, 5.1, 5.3, 5.9, 6.0, 6.8, 7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 10.3, 10.5])
y = np.array([39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 57189, 63218, 55794, 61111, 66029, 83088, 81363, 93940, 91738, 98273, 101302, 113812, 109431, 105582, 116969, 122391, 121872])

# Calculate the mean of X and y
mean_X = np.mean(X)
mean_y = np.mean(y)

# Calculate the slope (b1) and intercept (b0)
numerator = np.sum((X - mean_X) * (y - mean_y))
denominator = np.sum((X - mean_X) ** 2)
b1 = numerator / denominator
b0 = mean_y - b1 * mean_X

print(f"Slope (b1): {b1}")
print(f"Intercept (b0): {b0}")

# Predict y values
y_pred = b0 + b1 * X

# Calculate SSE, SST, and SSR
sse = np.sum((y - y_pred) ** 2)  # Sum of Squares for Error
sst = np.sum((y - mean_y) ** 2)  # Total Sum of Squares
ssr = sst - sse  # Sum of Squares for Regression

# R-squared
r2 = 1 - (sse / sst)

print(f"SSE (Sum of Squares for Error): {sse}")
print(f"SST (Total Sum of Squares): {sst}")
print(f"SSR (Sum of Squares for Regression): {ssr}")
print(f"R-squared: {r2}")

# Visualization
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Hard-Coded Linear Regression')
plt.legend()
plt.show()

```
